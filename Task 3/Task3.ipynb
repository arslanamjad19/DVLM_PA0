```python
import os
import math
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, utils
import matplotlib.pyplot as plt
from tqdm import tqdm

# -------------------------
# Config / Hyperparameters
# -------------------------
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

latent_dim = 2
batch_size = 128
epochs = 30
learning_rate = 1e-3
kl_anneal_epochs = 10  # number of epochs to ramp KL weight from 0 to 1
reconstruction_loss_reduction = 'sum'  # mirror Keras behaviour: sum across pixels then mean across batch

data_dir = "./data"
os.makedirs(data_dir, exist_ok=True)

# -------------------------
# Dataset and DataLoaders
# -------------------------
transform = transforms.Compose([
    transforms.ToTensor(),  # gives range [0,1]
])

full_train = datasets.FashionMNIST(root=data_dir, train=True, download=True, transform=transform)
test_set = datasets.FashionMNIST(root=data_dir, train=False, download=True, transform=transform)

# optional small validation split from train
val_size = 5000
train_size = len(full_train) - val_size
train_set, val_set = random_split(full_train, [train_size, val_size], generator=torch.Generator().manual_seed(seed))

train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True, pin_memory=True)
val_loader = DataLoader(val_set, batch_size=batch_size, shuffle=False, pin_memory=True)
test_loader = DataLoader(test_set, batch_size=batch_size, shuffle=False, pin_memory=True)

# -------------------------
# Model (encoder, decoder)
# -------------------------
class Sampling(nn.Module):
    def forward(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

class Encoder(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        # Input: (B, 1, 28, 28)
        self.conv1 = nn.Conv2d(1, 64, kernel_size=3, stride=2, padding=1)  # -> (B,64,14,14)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1)  # -> (B,128,7,7)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(128 * 7 * 7, 16)
        self.fc_mu = nn.Linear(16, latent_dim)
        self.fc_logvar = nn.Linear(16, latent_dim)
        self.act = nn.ReLU()

    def forward(self, x):
        x = self.act(self.conv1(x))
        x = self.act(self.conv2(x))
        x = self.flatten(x)
        x = self.act(self.fc1(x))
        mu = self.fc_mu(x)
        logvar = self.fc_logvar(x)
        return mu, logvar

class Decoder(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.fc = nn.Linear(latent_dim, 7 * 7 * 64)
        self.deconv1 = nn.ConvTranspose2d(64, 128, kernel_size=3, stride=2, padding=1, output_padding=1)  # -> (B,128,14,14)
        self.deconv2 = nn.ConvTranspose2d(128, 64, kernel_size=3, stride=2, padding=1, output_padding=1)  # -> (B,64,28,28)
        self.out = nn.ConvTranspose2d(64, 1, kernel_size=3, padding=1)  # -> (B,1,28,28)
        self.act = nn.ReLU()
        # final activation -> sigmoid for [0,1] output
    def forward(self, z):
        x = self.fc(z)
        x = x.view(-1, 64, 7, 7)
        x = self.act(self.deconv1(x))
        x = self.act(self.deconv2(x))
        x = torch.sigmoid(self.out(x))
        return x

class VAE(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.encoder = Encoder(latent_dim)
        self.decoder = Decoder(latent_dim)
        self.sampler = Sampling()

    def reparameterize(self, mu, logvar):
        return self.sampler(mu, logvar)

    def forward(self, x):
        mu, logvar = self.encoder(x)
        z = self.reparameterize(mu, logvar)
        recon = self.decoder(z)
        return recon, mu, logvar, z

# create model
vae = VAE(latent_dim).to(device)

# -------------------------
# Loss components
# -------------------------
def compute_losses(recon_x, x, mu, logvar, reduction='mean_batch'):
    """
    recon_x: (B,1,28,28) predicted in [0,1]
    x: (B,1,28,28)
    mu, logvar: (B, latent_dim)
    returns: reconstruction_loss (scalar), kl_loss (scalar), total (scalar)
    """
    batch_size = x.shape[0]
    # Reconstruction loss: MSE, sum across pixels for each sample, then mean across batch (to mirror Keras)
    # Equivalent to: reduce_sum(binary_crossentropy, axis=(1,2)) then mean in Keras example.
    # For MSE:
    recon_loss = F.mse_loss(recon_x, x, reduction='sum') / batch_size  # scalar
    # KL divergence (analytical): sum across latent dim per sample, then mean across batch
    # KL = -0.5 * sum(1 + logvar - mu^2 - exp(logvar))
    kl_element = 1 + logvar - mu.pow(2) - logvar.exp()
    kl_loss = -0.5 * torch.sum(kl_element) / batch_size
    total_loss = recon_loss + kl_loss
    return recon_loss, kl_loss, total_loss

# -------------------------
# Training function
# -------------------------
optimizer = torch.optim.Adam(vae.parameters(), lr=learning_rate)

def kl_weight_for_epoch(epoch, total_anneal_epochs=kl_anneal_epochs):
    # linear annealing from 0 -> 1 over total_anneal_epochs
    if total_anneal_epochs <= 0:
        return 1.0
    return float(min(1.0, epoch / total_anneal_epochs))

def train_epoch(model, loader, optimizer, epoch, kl_anneal_epochs):
    model.train()
    running_recon = 0.0
    running_kl = 0.0
    running_total = 0.0
    n_batches = 0
    kl_weight = kl_weight_for_epoch(epoch, kl_anneal_epochs)

    pbar = tqdm(loader, desc=f"Train E{epoch+1} KLw{kl_weight:.3f}", leave=False)
    for batch in pbar:
        x, _ = batch
        x = x.to(device)
        optimizer.zero_grad()
        recon, mu, logvar, _ = model(x)
        recon_loss, kl_loss, total_loss = compute_losses(recon, x, mu, logvar)
        weighted_total = recon_loss + kl_weight * kl_loss
        weighted_total.backward()
        optimizer.step()

        running_recon += recon_loss.item()
        running_kl += kl_loss.item()
        running_total += weighted_total.item()
        n_batches += 1
        pbar.set_postfix({"recon": running_recon / n_batches, "kl": running_kl / n_batches, "tot": running_total / n_batches})

    return {
        "reconstruction_loss": running_recon / n_batches,
        "kl_loss": running_kl / n_batches,
        "total_loss": running_total / n_batches,
        "kl_weight": kl_weight,
    }

def validate_epoch(model, loader, epoch, kl_anneal_epochs):
    model.eval()
    running_recon = 0.0
    running_kl = 0.0
    running_total = 0.0
    n_batches = 0
    kl_weight = kl_weight_for_epoch(epoch, kl_anneal_epochs)
    with torch.no_grad():
        for batch in loader:
            x, _ = batch
            x = x.to(device)
            recon, mu, logvar, _ = model(x)
            recon_loss, kl_loss, total_loss = compute_losses(recon, x, mu, logvar)
            weighted_total = recon_loss + kl_weight * kl_loss

            running_recon += recon_loss.item()
            running_kl += kl_loss.item()
            running_total += weighted_total.item()
            n_batches += 1

    return {
        "reconstruction_loss": running_recon / n_batches,
        "kl_loss": running_kl / n_batches,
        "total_loss": running_total / n_batches,
        "kl_weight": kl_weight,
    }

# -------------------------
# Training loop
# -------------------------
train_logs = {"train":[], "val":[]}

for epoch in range(epochs):
    train_stats = train_epoch(vae, train_loader, optimizer, epoch, kl_anneal_epochs)
    val_stats = validate_epoch(vae, val_loader, epoch, kl_anneal_epochs)
    train_logs["train"].append(train_stats)
    train_logs["val"].append(val_stats)
    # print epoch summary
    print(f"Epoch {epoch+1}/{epochs} | Train tot: {train_stats['total_loss']:.4f}, recon: {train_stats['reconstruction_loss']:.4f}, kl: {train_stats['kl_loss']:.4f} | Val tot: {val_stats['total_loss']:.4f}, recon: {val_stats['reconstruction_loss']:.4f}, kl: {val_stats['kl_loss']:.4f} | KLw: {train_stats['kl_weight']:.3f}")

# -------------------------
# Utility: plot losses
# -------------------------
def plot_training_curves(logs):
    epochs = len(logs["train"])
    train_recon = [logs["train"][i]["reconstruction_loss"] for i in range(epochs)]
    train_kl = [logs["train"][i]["kl_loss"] for i in range(epochs)]
    train_total = [logs["train"][i]["total_loss"] for i in range(epochs)]
    val_recon = [logs["val"][i]["reconstruction_loss"] for i in range(epochs)]
    val_kl = [logs["val"][i]["kl_loss"] for i in range(epochs)]
    val_total = [logs["val"][i]["total_loss"] for i in range(epochs)]
    kl_weights = [logs["train"][i]["kl_weight"] for i in range(epochs)]

    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.plot(train_total, label='train total')
    plt.plot(val_total, label='val total')
    plt.title('Total loss')
    plt.legend()
    plt.subplot(1,3,2)
    plt.plot(train_recon, label='train recon')
    plt.plot(val_recon, label='val recon')
    plt.title('Reconstruction loss (MSE)')
    plt.legend()
    plt.subplot(1,3,3)
    plt.plot(train_kl, label='train kl')
    plt.plot(val_kl, label='val kl')
    plt.title('KL loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_training_curves(train_logs)

# -------------------------
# Reconstructions (visualize)
# -------------------------
def visualize_reconstructions(model, dataloader, n=10):
    model.eval()
    images = []
    originals = []
    with torch.no_grad():
        for x, _ in dataloader:
            originals.append(x)
            x = x.to(device)
            recon, _, _, _ = model(x)
            images.append(recon.cpu())
            if len(images) >= 1:
                break
    originals = torch.cat(originals, dim=0)[:n]
    recons = torch.cat(images, dim=0)[:n]
    # display side-by-side
    comparison = torch.cat([originals, recons])
    grid = utils.make_grid(comparison, nrow=n, pad_value=1.0)
    plt.figure(figsize=(12,4))
    plt.imshow(grid.permute(1,2,0).squeeze(), cmap='gray')
    plt.axis('off')
    plt.title("Top: originals, Bottom: reconstructions")
    plt.show()

visualize_reconstructions(vae, test_loader, n=10)

# -------------------------
# Generations from priors
# -------------------------
def generate_from_prior(model, prior='gaussian', n=25, figsize=(6,6)):
    model.eval()
    if prior == 'gaussian':
        z = torch.randn(n, latent_dim).to(device)
    elif prior == 'laplace':
        # sample Laplace(0,1) using torch.distributions
        lap = torch.distributions.Laplace(loc=0.0, scale=1.0)
        z = lap.sample((n, latent_dim)).to(device)
    else:
        raise ValueError("Unsupported prior")
    with torch.no_grad():
        samples = model.decoder(z).cpu()
    grid = utils.make_grid(samples, nrow=int(math.sqrt(n)), pad_value=1.0)
    plt.figure(figsize=figsize)
    plt.imshow(grid.permute(1,2,0).squeeze(), cmap='gray')
    plt.axis('off')
    plt.title(f"Samples from prior: {prior}")
    plt.show()

# generate from Gaussian prior
generate_from_prior(vae, prior='gaussian', n=25)

# generate from Laplacian prior
generate_from_prior(vae, prior='laplace', n=25)

# -------------------------
# Latent scatter colored by labels
# -------------------------
def plot_latent_clusters(encoder, dataloader, n_samples=2000):
    encoder.eval()
    zs = []
    labels = []
    with torch.no_grad():
        for x, y in dataloader:
            x = x.to(device)
            mu, logvar = encoder(x)
            zs.append(mu.cpu().numpy())
            labels.append(y.numpy())
            if sum(len(a) for a in labels) >= n_samples:
                break
    zs = np.concatenate(zs, axis=0)[:n_samples]
    labels = np.concatenate(labels, axis=0)[:n_samples]
    plt.figure(figsize=(8,6))
    sc = plt.scatter(zs[:,0], zs[:,1], c=labels, cmap='tab10', s=6)
    cbar = plt.colorbar(sc, ticks=range(10))
    cbar.ax.set_yticklabels(['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot'])
    plt.xlabel('z[0]'); plt.ylabel('z[1]'); plt.title('Latent means colored by label')
    plt.show()

plot_latent_clusters(vae.encoder, test_loader)

# -------------------------
# Save model and logs
# -------------------------
torch.save({
    'model_state_dict': vae.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'train_logs': train_logs
}, "vae_fashionmnist.pth")
print("Saved checkpoint: vae_fashionmnist.pth")
```
